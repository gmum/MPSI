# Zadanie: Klasyfikacja ryzyka zawału serca

Twoim zadaniem jest zbudowanie prostego klasyfikatora, który będzie przewidywał ryzyko zawału serca na podstawie zestawu cech pacjenta.  
Celem jest **optymalizacja metryk klasyfikacji** (`precision`, `recall`, `accuracy`) w zależności od przyjętej strategii.

---

## Dane

Plik: `heart_disease_balanced.csv`  
Każdy wiersz reprezentuje pacjenta, a kolumny to cechy kliniczne:
- `age` – wiek  
- `sex` – płeć (0: kobieta, 1: mężczyzna)  
- `resting_bp` – ciśnienie spoczynkowe  
- `cholesterol` – poziom cholesterolu  
- `max_heart_rate` – maksymalne tętno  
- `exercise_angina` – ból w klatce przy wysiłku  
- `oldpeak` – depresja ST  
- `target` – (0: brak zawału, 1: zawał)  

Zbiór jest **zbalansowany** (po 250 przykładów z każdej klasy).

## Część 3 – Optymalizacja

Załóż dwa scenariusze:

### Scenariusz A – Minimalizujemy fałszywe alarmy
- Celem jest **maksymalizacja precision**

### Scenariusz B – Wykrywamy jak najwięcej zawałów
- Celem jest **maksymalizacja recall**

Dla każdego scenariusza:
- Dobierz optymalny **próg klasyfikacji**.
- Porównaj metryki (`accuracy`, `precision`, `recall`, `F1`).
- Oceń kompromisy między nimi.
